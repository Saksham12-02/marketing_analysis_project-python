```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Marketing_Analysis.csv')
```

```python
df = pd.read_csv('Marketing_Analysis.csv', skiprows = 2)
```

```python
df
```

```python
df['job'] = df['jobedu'].apply(lambda x : x.split(',')[0])
```

```python
df['education'] = df['jobedu'].apply(lambda x : x.split(',')[1])
```

```python
df.drop('jobedu', axis = 1, inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
mean_age = df['age'].mean()
df['age'].fillna(value = mean_age, inplace = True)
df.isnull().sum()
```

```python
mode_month = df['month'].mode()
df['month'].fillna(value = mode_month[0], inplace = True)
df.isnull().sum()
```

```python
df = df[~df.response.isnull()].copy()
df.isnull().sum()
```

```python
df
```

```python
c = df.job.value_counts()
c
```

```python
c = df.job.value_counts(normalize = True)
c
```

```python
df.job.value_counts(normalize = True).plot(kind='pie', autopct='%1.2f%%')
plt.show()
```

```python
df.plot.scatter(x = 'age', y = 'salary' )
plt.show()
```

```python
df.plot.scatter(x = 'balance', y = 'salary' )
plt.show()
```

```python
plt.scatter(df.age, df.balance, marker = '^')
plt.scatter(df.salary, df.balance, marker = '*')
plt.show()
```

```python
sns.pairplot(data = df, vars =['salary','balance','age'])
plt.show()
```

```python
sns.heatmap(df[['age', 'salary', 'balance']].corr(),annot= True, cmap = 'Reds')
plt.show()
```

```python
print(df.groupby('response')['salary'].mean())
print(df.groupby('response')['salary'].median())
plt.show()
```

```python
sns.boxplot(x = df.response, y = df.salary)
plt.show()
```

```python
df['response_rate'] = np.where(df['response'] == 'yes', 1, 0)
df.response_rate.value_counts()
```

```python
df.groupby('marital')['response_rate'].plot.barh()
plt.show()
```
